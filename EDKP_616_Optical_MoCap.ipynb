#### Multi-segment foot model basics: ####

Gait assessment has been extensively used in the literature for both clinical and industrial applications. As such, much work has gone into understanding and modelling the kinematic behaviour of the foot during gait. At large, foot kinematics can be measured with varying levels of accuracy depending on the goal of the investigation.  For example, many investigations choose to use a simplified, one-segment model of the foot for gait assessment (e.g., the plug-in-gait model (PiG)) due to its ease of use and lower computational burden ([Kabata et al., 1990](https://doi.org/10.1002/jor.1100080310); [Davis et al., 1991](https://doi.org/10.1016/0167-9457(91)90046-Z)). Alternatively, the Oxford Foot Model (OFM) and other multi-segment foot models (Rizzoli Foot Model, Amsterdam Foot Model, etc.) have been widely used in investigations which require the quantification of the foot’s different biomechanical segments ([Carson et al., 2001](https://doi.org/10.1016/S0021-9290(01)00101-4); [Dixon et al., 2025](https://doi.org/10.1080/10255842.2024.2448558); [Leardini et al., 2019]( https://doi.org/10.1016/j.gaitpost.2019.01.022 )). Although multi-segment foot models are robust, their implementation can be laborious, requiring substantially more markers than single-segment counterparts (Table 1). This additional complexity can discourage its adoption.

Intuitively, it is hard to imagine that a single-segment approach would have high validity when trying to track the motion of a segment that is comprised of 28 bones, 33 joints, 112 ligaments, and is controlled by approximately 34 muscles. Indeed, recent work has highlighted the negative repercussions of such simplifications, as the PiG model produces opposite kinematic results at the ankle joint when compared to the OFM ([Pothrat et al., 2015](https://doi.org/10.1016/j.clinbiomech.2015.03.004)).

The OFM is a leading multi-segment foot model, initially developed by [Carson et al. (2001)](https://doi.org/10.1016/S0021-9290(01)00101-4) to provide a thorough and valid standardized testing protocol. The implementation of the OFM directly into the population Vicon Nexus software (Vicon Motion Systems Ltd., Oxford, UK) can be largely credited with the model’s popularity. The OFM tracks the foot’s three main biomechanical segments (i.e. the hindfoot, forefoot, and hallux) as well as the tibia. The model then outputs the segmental angles in the sagittal, coronal, and transverse planes. Since its original implementation has been extensively improved upon. For example, the model has been assessed for repeatability in different populations, such as children ([Stebbins et al., 2006](https://doi.org/10.1016/j.gaitpost.2005.03.002)). Furthermore, a recent project from the MOTION lab aimed at providing an open-source implementation of the OFM (termed “openOFM”) to provide transparency into the model’s calculations ([Dixon et al., 2025](https://doi.org/10.1080/10255842.2024.2448558)).

#### Clinical applications of multi-segment foot models: ####

Clinical applications of multi-segment foot models are widespread, with their use being particularly impactful in the domains of pathological gait diagnosis, assessment of surgical outcomes, and comparison between surgical techniques ([Leardini et al., 2019]( https://doi.org/10.1016/j.gaitpost.2019.01.022 )). Accordingly, the development of repeatable, valid, and accessible multi-segment foot models directly aims to improve clinical care and improvements in rehabilitation. Work by the Oxford Foot Laboratory demonstrates a research program which meaningfully applied the OFM clinically. Work from the group has used the OFM to identify relationships between flat feet and knee, hip, and back discomfort ([Kothari et al., 2016](https://doi.org/10.1016/j.gaitpost.2016.02.008)), identify links between foot function and quality of life ([Kothari et al., 2015](https://doi.org/10.1016/j.gaitpost.2015.02.012)), and characterize proximal gait compensations in children with cerebral palsy ([Stebbins et al., 2010](https://doi.org/10.1016/j.gaitpost.2010.05.006)).

Like any other methodology, current optical motion capture-driven multi-segment foot models, such as the OFM, are not without limitations. For example, their reliance on the placement of markers on specific anatomical landmarks makes them susceptible to experimenter expertise and consistency. Furthermore, the OFM currently requires optical motion capture systems, which can be incredibly expensive. The costly equipment, as well as the relatively long time needed to place anatomical markers, render the OFM inaccessible in many clinical settings. Advances in the accessibility of such models would allow for increased access to such a critical clinical tool.

#### Using optical markers to track segment orientation: ####

In optical motion capture-based systems, the calculation of segment kinematics is achieved by tracking the position of markers in 3D space. More specifically, participants will start by completing a static calibration trial during which they have markers on specific anatomical landmarks and one rigid body per segment. Calibration trials establish the position of the anatomical landmarks in relation to the tracking clusters and allow for the anatomic landmark markers to be removed and be tracked solely using the rigid bodies. This is particularly useful when markers on anatomical landmarks could impede normal movement patterns. For example, markers on the medial and lateral epicondyles would be particularly prone to dislodgement during a gait cycle.

Based on these markers, local coordinate systems are constructed for each segment, from which inter-segmental angles can be derived. The openOFM employs the Joint Coordinate System framework proposed by [Grood and Suntay (1983)](https://doi.org/10.1115/1.3138397) to quantify relative segmental orientations. A sample workflow outlining how inter-segment angles can be calculated is provided below.

#### Solving for orientation using the joint coordinate system ####

When using the [Grood and Suntay (1983)](https://doi.org/10.1115/1.3138397) method for calculating joint angles, one must first start by creating the local coordinate systems for both the proximal and the distal segment. For this example, the conventions introduced by [Grood and Suntay (1983)](https://doi.org/10.1115/1.3138397) for the knee joint will be followed directly for the angle between the proximal tibia and distal hindfoot. Therefore, a coordinate system with I pointing laterally, J pointing anteriorly, and K pointing proximally must be constructed.

For the following calculations, markers for the proximal and distal segments are labelled as follows:


<center>
<img src="figures/marker_labels.png" alt="Marker Labels" width="600">
</center>
<p><strong>Figure 1.</strong> Visualization of the local coordinate systems and marker labels for both the proximal (tibia) and distal (hindfoot) segments.</p>
<br>

Firstly, a unit vector on the proximal segment is made to represent K:


<br>
<center>
<br>
$\displaystyle\ {K} = \frac{M2-M1}{\|M2-M1\|}$
</center>
<br>

Next, a temporary vector is made in the medial/lateral axis:
<br>
<center>
$\displaystyle\ {V_{temp}} = \frac{M3-M1}{\|M3-M1\|}$
</center>
<br>

Following this, a unit vector (J) can be made which is perfectly orthogonal to K:
<br>
<center>
$\displaystyle\ {J} = \frac{K \times V}{\|K \times V\|}$
</center>
<br>

Finally, a unit vector (I) can be produced:
<br>
<center>
$\displaystyle\ {I} = \frac{J \times K}{\|J \times K\|}$
</center>
<br>

Now a right handed, orthogonal coordinate system is created (Fig. 1). The same steps are repeated using the corresponding marker numbers for the distal segment. Right handedness and orthogonality can be verified by making sure that the coordinate system has a determinant of +1.0.

The axes that are the most clinically meaningful to the joint must be identified to determine the optimal rotation order. For the purpose of this example, the flexion/extension and internal/external rotation of the segment will be used as the segment fixed axes for the proximal and distal segments, respectively. Accordingly, the adduction/abduction of the joint is the least clinically meaningful and would correspond to the second rotation in the corresponding Euler sequence.

Therefore, $e_1$ is defined as the medial/lateral axis of the proximal segment (tibia) and $e_3$ is defined as the longitudinal axis of the distal segment (hindfoot). The 'floating axis'/ $e_2$ is then defined by taking the cross product between $e_3$ and $e_1$. This is the axis about which abduction/adduction of the joint is about.

The axes discussed above are visualized in the following image:


<center>
<img src="figures/g_and_s.png" alt="JCS Figure" width="600">
</center>

Figure 2. Visualization of the axes definitions and angles defined used in the joint coordinate system presented by [Grood & Suntay (1983)](https://doi.org/10.1115/1.3138397). The figure has been adapted from [Nigg & Herzog (2007)](https://courses.washington.edu/bioen520/notes/Nigg-Joint_Coordinate_Systems.pdf).

It is worth noting that alpha, beta, and gamma are not fixed to defining flexion, adduction, or rotation, but rather, they describe rotation about e1, e2, and e3. So, the movement that they describe can change depending on the axes that are defined as such (i.e., the rotation order that is chosen).

<br>

Now, $\alpha$, $\beta$, and $\gamma$ can be isolated using the following equations provided by [Grood & Suntay (1983)](https://doi.org/10.1115/1.3138397):



<center>
$\displaystyle \sin(\alpha) = -\mathbf{e}_2 \cdot \mathbf{K}$
</center>
<br>
<center>
$\displaystyle \cos(\beta) = I \cdot k$
</center>
<br>
<center>
$\displaystyle \sin(\gamma) = -\mathbf{e}_2 \cdot \mathbf{i}$
</center>
<br>

### Sample Code: The Joint Coordinate System ###

Methods similar to the Grood & Suntay calculations above were used in the OpenOFM ([Dixon et al., 2025](https://doi.org/10.1080/10255842.2024.2448558)). This code has recently been made available on a [public repository](https://github.com/mcgillmotionlab/openOFM).

```python
from mocap_functions import load, proximal_local_cord_sys, distal_local_cord_sys, grood_angles, plot

# Loading data:
data = load('data/dynamic01.c3d')

# Build the local coordinate systems
J, K, I = proximal_local_cord_sys(data['M1'], data['M2'], data['M3'])
j, k, i = distal_local_cord_sys(data['m1'], data['m3'], data['m4'])

# Calculate joint angles
alpha, beta, gamma, e2 = grood_angles(J, K, I, j, k, i)

# Visualize the data
plot(alpha, beta, gamma)
```
